# Visualization

```python
from ipywidgets import Label, HTML, HBox, Image, VBox, Box, HBox, Output, Layout, SelectMultiple, Button, ToggleButtons, Tab,Text
from ipyevents import Event 
from IPython.display import display
```

```python
def get_selected_data_type(data_type):
    if data_type == 'Categorical':
        return DataType.CATEGORICAL
    return DataType.CONTINUOUS

data_type_toggle = ToggleButtons(
    options=['Continuous', 'Categorical']
)

default_data_type = get_selected_data_type(data_type_toggle.value)

measurements = SelectMultiple(
    options=varcop.get_columns_for_type(default_data_type),
    value=varcop.get_columns_for_type(default_data_type),
    rows=30,
    layout=Layout(max_width='150px')
    
)
select_all_measurements = Button(
    description='Select All',
    icon='check'
)

graph = HTML(varcop.visualize(attributes=measurements.value), data_type=default_data_type)
info = Output()
log = HTML()
plot_tabs = Tab(layout={'display': 'none'})

events = Event(source=graph, watched_events=['click'], watched_node_attributes=['data-type', 'data-id', 'data-id-left', 'data-id-right'])

@info.capture(clear_output=True)
def handle_graph_click_event(event):
    log.value = (f"You clicked a {event}")
    df_to_display = None
    plots_to_display = []
    if event['data-type'] == 'node':
        df_to_display = varcop.dataframe_for_node(event['data-id'], attributes=measurements.value, data_type=get_selected_data_type(data_type_toggle.value))
        plots_to_display = [varcop.plot_for_node_attribute(event['data-id'], attribute, data_type=get_selected_data_type(data_type_toggle.value)) for attribute in df_to_display.reset_index()['variable'].tolist()]
    if event['data-type'] == 'edge':
        df_to_display = varcop.dataframe_for_edge(event['data-id-left'], event['data-id-right'], attributes=measurements.value, data_type=get_selected_data_type(data_type_toggle.value))
        plots_to_display = [varcop.plot_for_edge_attribute(event['data-id-left'], event['data-id-right'], attribute, data_type=get_selected_data_type(data_type_toggle.value)) for attribute in df_to_display.reset_index()['variable'].tolist()]
    if df_to_display is not None:    
        plot_tabs.layout.display = 'block'
        with info:
            display(df_to_display)
        outputs = [Output() for plot in plots_to_display]
        for i, plot in enumerate(plots_to_display):
            with outputs[i]:
                display(plot)
        plot_tabs.children = outputs
        [plot_tabs.set_title(i, tab) for i, tab in enumerate(df_to_display.reset_index()['variable'].tolist())]
        
    
@info.capture(clear_output=True)
def handle_data_type_update_event(change):
    plot_tabs.layout.display = 'none'
    new_data_type = get_selected_data_type(change['new'])
    measurements.options = varcop.get_columns_for_type(new_data_type)
    measurements.value = varcop.get_columns_for_type(new_data_type)
    graph.value = varcop.visualize(attributes=measurements.value, data_type=new_data_type)
    
@info.capture(clear_output=True)
def handle_measurements_update_event(change):
    plot_tabs.layout.display = 'none'
    graph.value = varcop.visualize(attributes=change['new'], data_type=get_selected_data_type(data_type_toggle.value))
            
def handle_select_all_measurements(event):
    measurements.value = measurements.options

events.on_dom_event(handle_graph_click_event)
data_type_toggle.observe(handle_data_type_update_event, names='value')
measurements.observe(handle_measurements_update_event, names='value')
select_all_measurements.on_click(handle_select_all_measurements)

display(HBox([VBox([data_type_toggle,select_all_measurements,measurements]), VBox([info, plot_tabs, graph])]))
```
